# Lowest Risk Majors

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary']-clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1,'Spread',spread_col)
clean_df.head()
```

# Major with the Highest Potential

```python
heighest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary',ascending=False)
heighest_potential[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head()
```

# Major with lowest Spread or low risk majors

```python
lowest_spread = clean_df.sort_values('Spread',ascending=True) #by default ascending sort is true
lowest_spread[['Undergraduate Major','Spread']].head()
```

#Major with greatest Spread

```python
Greatest_Spread = clean_df.sort_values('Spread',ascending=False)
Greatest_Spread[['Undergraduate Major','Spread']].head()
```

```python
heighest_spread = clean_df.sort_values('Mid-Career Median Salary',ascending=False)
heighest_spread[['Undergraduate Major','Mid-Career Median Salary']].head()
```

```python
# Majors in each Category of group HASS STEM or Business
clean_df.groupby('Group').count()
```

```python
#Average salary by group
# to make values more readable
pd.options.display.float_format = '{:,.2f}'.format 
clean_df.groupby('Group').mean()
```
